```python
from selenium import webdriver
import time
import pandas as pd
from selenium.webdriver.common.by import By
```

```python
options = webdriver.ChromeOptions()
options.add_experimental_option('excludeSwitches', ['enable-logging'])
driver = webdriver.Chrome(options=options)

# create an empty list to store the news data
news_data = []
category = "Funding"

# loop through each page and extract the news data
for page in range(1, 16):  # change the range to extract more pages
    # navigate to the desired web page
    print(f"Extracting data from page {page}...")
    url = f'https://www.prnewswire.com/search/news/?keyword=funding&page={page}&pagesize=75'
    driver.get(url)

    # wait for the page to load
    time.sleep(2)

    # find all the news articles on the page
    articles = driver.find_elements(
        By.XPATH, '//div[@class="row arabiclistingcards"]/div[@class="card"]')

    # loop through each article and extract the required information
    for article in articles:
        print("Extracting data from article...")
        # extract the news title
        title = article.find_element(By.XPATH, './/h3/a').text

        # extract the news date
        date = article.find_element(By.XPATH, './/small').text

        # extract the news content
        article_body = article.find_element(By.XPATH, './/p').text

        # extract the news link
        link = article.find_element(By.XPATH, './/h3/a').get_attribute('href')

        # add the news data to the list
        news_data.append([title, date, article_body, link, category])

    print(f"Extracted data from all articles on page {page}.")

# create a pandas DataFrame from the news data
df = pd.DataFrame(news_data, columns=['News_Title', 'Date', 'Content', 'Link', 'category'])

# save the DataFrame to an csv file
df.to_csv('final_funding.csv', index=False)

# close the browser window
driver.quit()
```
