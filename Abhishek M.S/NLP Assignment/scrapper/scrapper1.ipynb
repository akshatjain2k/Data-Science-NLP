```python
# Importing Important Libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
from googletrans import Translator
```

```python
date_ = []
title_ = []
more_info = []
body_news = []
company_name = []

try:
    url = requests.get("https://www.businesswire.com/portal/site/home/template.PAGE/news/subject/?javax.portlet.tpst=08c2aa13f2fe3d4dc1b6751ae1de75dd&javax.portlet.prp_08c2aa13f2fe3d4dc1b6751ae1de75dd_vnsId=31333&javax.portlet.prp_08c2aa13f2fe3d4dc1b6751ae1de75dd_viewID=MY_PORTAL_VIEW&javax.portlet.prp_08c2aa13f2fe3d4dc1b6751ae1de75dd_ndmHsc=v2*A1677416400000*B1679997791327*DgroupByDate*G3*M31333*N1000105&javax.portlet.begCacheTok=com.vignette.cachetoken&javax.portlet.endCacheTok=com.vignette.cachetoken").text
except:
    print("Link not working start page")
    
soup = BeautifulSoup(url)
news = soup.find_all("li", class_ = "")

for new in news:
    title = new.find("a", class_ = "bwTitleLink")
    link = new.find('a')['href']
    date = new.find("div", class_ = "bwTimestamp")
    
    
    ### for title 
    try:             
        title_text = title.text
        
        translator = Translator()
        translated = translator.translate(title_text, dest = 'en')
        title_.append(translated.text)
        
        #print(title_text)
        
    except:
        title_.append("None")
        
        
        
    ### for links    
    try:             
        
        l = f"https://www.businesswire.com{link}"
        more_info.append(l)
        #print(l)
        
        
    except:
        more_info.append("None")
        
        
    ### for date
    try:             
        
        d = date.text
        
        date_.append(d[17:38])
        #print(d)
        
    except:
        date_.append("None")
        
        
    ### code for the next page deatiled news
    try:    
        n_url = url = requests.get(l).text
    except:
        print("Link not working for next page")
        
    soup2 = BeautifulSoup(n_url)
    body = soup2.find("p", class_ = "")
    
    ### for body of the news in next page
    try:
        body_text = body.text
        
        translator = Translator()
        translated = translator.translate(body_text, dest = 'en')
        
        body_news.append(translated.text)
        #print(body.text)
    except:
        body_news.append("None")
    
```
