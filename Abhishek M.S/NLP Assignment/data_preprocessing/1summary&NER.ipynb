```python
# summarise the article_body

import nltk
from nltk.corpus import stopwords

nltk.download('punkt')
```

```python
# summarise the article_body
# Load Packages
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.text_rank import TextRankSummarizer


def summarize_data(text):
    parser = PlaintextParser.from_string(text, Tokenizer("english"))
#     print(len(text))
    # Summarize using sumy TextRank
    summarizer = TextRankSummarizer()
    summary = summarizer(parser.document, 2)
    text_summary = ""
    for sentence in summary:
      text_summary += str(sentence)

#     print(len(text_summary))
    return text_summary
```

```python
sum_data = []
for x in range(0, len(df)):
  if type(df['article_body'][x]) == str:
    summary_output = summarize_data(df['article_body'][x])
    ##print(summary_output)
    ##print(df['Full_News'][x])
    sum_data.append(summary_output)
  else:
    sum_data.append("NaN")


df['summary'] = sum_data
df.head()
```

```python
# finding company name using NER

import spacy

# Load the pre-trained NER model from spaCy
nlp = spacy.load('en_core_web_sm')
```

```python
# Define a function to extract ORG entities from a text using spaCy
def company_name(text):
    doc = nlp(text)
    orgs = [entity.text for entity in doc.ents if entity.label_ == 'ORG']
    return orgs


org_data = []
for x in range(0, len(df)):
  if type(df['summary'][x]) == str:
    org_output = company_name(df['summary'][x])
    ##print(summary_output)
    ##print(df['Full_News'][x])
    org_data.append(org_output)
  else:
    org_data.append("NaN")
    
df['company_name'] = org_data
df.head()
```

```python
# Define a function to extract amount/money/price entities from a text using spaCy

def funds(text):
    doc = nlp(text)
    money = [entity.text for entity in doc.ents if entity.label_ == 'MONEY']
    return money


funds_data = []
for x in range(0, len(df)):
    if type(df['article_body'][x]) == str:
        funds_output = funds(df['article_body'][x])
        ##print(summary_output)
        ##print(df['Full_News'][x])
        funds_data.append(funds_output)
    else:
        funds_data.append("NaN")


df['funds'] = funds_data
df.head()
```
