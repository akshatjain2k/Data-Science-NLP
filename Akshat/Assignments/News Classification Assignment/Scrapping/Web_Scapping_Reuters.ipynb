```python
import requests
from bs4 import BeautifulSoup

# Set the URL to scrape
url = 'https://www.reuters.com/news/archive/businessnews?view=page&page={}'

# Set the number of pages to scrape
num_pages = 80

# Create an empty list to store the article text
titles = []
body = []
links = []
dates = []


# Loop through each page and scrape the article text
for page_num in range(1, num_pages+1):
    page_url = url.format(page_num)
    page = requests.get(page_url)
    soup = BeautifulSoup(page.content, 'html.parser')
    article_blocks = soup.find_all('article', class_='story')
    #links = soup.find_all('div', class_='story-photo lazy-photo')

    for block in article_blocks:
      link = block.find('a')['href']
      try:
         
        article = block.find('time').get_text()
        dates.append(article)
      except:
        dates.append("NA")

      try:
        article = block.find('h3').get_text()
        titles.append(article)
      except:
        titles.append("NA")

      try:
        article = block.find('p').get_text()
        body.append(article)
      except:
        body.append("NA")
      
      try:
        l = f"https://www.reuters.com{link}"
        links.append(l)
      except:
        links.append("NA")

# Print the number of articles collected
```

```python
print(len(titles))
print(len(body))
print(len(links))
print(len(dates))
```

```python
updates_titles = []
for t in titles:
   updates_titles.append(t[9:])
#updates_titles[1:5]
```

```python
updates_dates = []
for d in dates:
   updates_dates.append(d[1:12])
#updates_dates
```

```python
import pandas as pd
# Converting into DataFrame
dict_ = {"Date_Time":updates_dates, "Title":updates_titles, "Full_News":body, "Links": links, "Category": "Stock News"}
df = pd.DataFrame(dict_)
```

```python
df.head(50)
```

```python
# Replace the random digits and colons with a blank space
df['Date_Time'] = df['Date_Time'].replace('\d{1,2}:\d{2}\w{2} \w{3}\\n', 'Mar 29 2023', regex=True)
df['Date_Time'] = df['Date_Time'].replace('\d{1,2}:\d{2}\w{2} \w{3}', 'Mar 29 2023', regex=True)
df['Date_Time'] = df['Date_Time'].replace('NA', 'NaN')


# Print the updated DataFrame
df.head(50)
```

```python
df.dropna(inplace = True)
```
