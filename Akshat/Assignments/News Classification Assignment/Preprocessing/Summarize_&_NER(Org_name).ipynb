```python
import pandas as pd
df = pd.read_csv('Final_Stock_news.csv')
df.head()
```

```python
import nltk
nltk.download('punkt')
  
```

## Summarize text

```python
# Load Packages
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.text_rank import TextRankSummarizer


def summarize_data(text):
    parser = PlaintextParser.from_string(text, Tokenizer("english"))
    print(len(text))
    # Summarize using sumy TextRank
    summarizer = TextRankSummarizer()
    summary = summarizer(parser.document, 2)
    text_summary = ""
    for sentence in summary:
      text_summary += str(sentence)
    
    print(len(text_summary))
    return text_summary
    
```

## Apply in CSV File Column

```python
sum_data = []
for x in range(0, len(df)):
  if type(df['Full_News'][x]) == str:
    summary_output = summarize_data(df['Full_News'][x])
    ##print(summary_output)
    ##print(df['Full_News'][x])
    sum_data.append(summary_output)
  else:
    sum_data.append("NaN")
```

```python
df['Summarize_data'] = sum_data
df.head()
```

## Finding Company Name using NER

```python
import spacy

# Load the pre-trained NER model from spaCy
nlp = spacy.load('en_core_web_sm')
```

```python
# Define a function to extract ORG entities from a text using spaCy
def company_name(text):
    doc = nlp(text)
    orgs = [entity.text for entity in doc.ents if entity.label_ == 'ORG']
    return orgs
```

## Apply in CSV File Column

```python
org_data = []
for x in range(0, len(df)):
  if type(df['Summarize_data'][x]) == str:
    org_output = company_name(df['Summarize_data'][x])
    ##print(summary_output)
    ##print(df['Full_News'][x])
    org_data.append(org_output)
  else:
    org_data.append("NaN")
```

```python
df['Company_Name'] = org_data
df.head()
```

```python
# save dataframe
# Summarize and company name added
df.to_csv('Updates_Stock_news.csv', index=False)
```
