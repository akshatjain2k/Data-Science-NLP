```python
import requests
from bs4 import BeautifulSoup
```

```python
def scrapper():
    ques = []
    ans = []
    for i in range(1,12):
            try:
                response = requests.get(f"https://www.bajajfinserv.in/contact-us-faq-queries")
            except:
                print("link not working")
            soup = BeautifulSoup(response.content, 'html.parser')
            news = soup.find_all("div", class_ = "accordion__row")
            for new in news:
                questions = new.find("div", class_ = "accordion__toggle accordion__title")
                answer = new.find("p")
                # for questions
                try:             
                    quesbody = questions.text

                    ques.append(quesbody)
                    
                    #print(quesbody)
                    
                except:
                    ques.append("None")

                ### for answers 
                try:             
                    ansbody = answer.text

                    ans.append(ansbody)
                    
                   # print(ansbody)
                    
                except:
                    ans.append("None")
               
    return ques, ans
```

```python
ques, ans = scrapper()
```

```python
print(len(ques))
print(len(ans))
```

```python
# Converting into DataFrame
import pandas as pd
dict_ = {"Questions":ques, "Answers":ans, "Category": "FAQ"}
df = pd.DataFrame(dict_)
df.head()
```

```python
# Removing extra space using re
import re
# Define a function to remove extra white space from a string
def clean_string(s):
    return re.sub('\s+', ' ', s).strip()
```

```python
clean_ques = []
for x in range(0, len(df)):
  if type(df['Questions'][x]) == str:
    regex_output = clean_string(df['Questions'][x])
    ##print(summary_output)
    ##print(df['Full_News'][x])
    clean_ques.append(regex_output)
  else:
    clean_ques.append("NaN")
```

```python
df['Question'] = clean_ques
df.head()
```

```python
# Deleting uncleaned column
df = df.drop('Questions', axis=1)
df.head()
```

```python
# Rearrange the columns in order
df = df[['Question', 'Answers', 'Category']]
df.head()
```

```python
# save dataframe
df.to_csv('FAQ.csv', index=False)
```
